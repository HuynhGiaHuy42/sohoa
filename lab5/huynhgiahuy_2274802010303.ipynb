```python
#bài tập về nhà
import pandas as pd

# Đọc dữ liệu
file_path = "howlongwelive.csv"
df = pd.read_csv(file_path)

# 2.1 Hiển thị số hàng và số cột
num_rows, num_cols = df.shape
print("=" * 50)
print(f" Thống kê dữ liệu:")
print(f"- Số hàng: {num_rows}")
print(f"- Số cột: {num_cols}")
print("=" * 50)

# 2.2 Danh sách tên cột
print("\n Danh sách các cột:")
print(df.columns.tolist())
print("=" * 50)

# 2.3 Thử truy xuất dữ liệu cột 'Status'
if "Status" in df.columns:
    print("\n Một số giá trị trong cột 'Status':")
    print(df["Status"].head())
else:
    print("\n Cột 'Status' không tồn tại trong dữ liệu!")
print("=" * 50)

# 2.4 Đếm số lượng nước đang phát triển và phát triển
if "Status" in df.columns:
    developing_count = (df["Status"] == "Developing").sum()
    developed_count = (df["Status"] == "Developed").sum()
    print("\n Phân loại quốc gia theo tình trạng phát triển:")
    print(f"- Số nước đang phát triển: {developing_count}")
    print(f"- Số nước phát triển: {developed_count}")
else:
    print("\n Không thể thực hiện thống kê vì thiếu cột 'Status'!")
print("=" * 50)

# 2.5 Thống kê các giá trị của 'Alcohol'
if "Alcohol" in df.columns:
    alcohol_min = df["Alcohol"].min()
    alcohol_max = df["Alcohol"].max()
    alcohol_mean = df["Alcohol"].mean()
    alcohol_median = df["Alcohol"].median()
    
    print("\n Thống kê mức tiêu thụ rượu (Alcohol):")
    print(f"- Min: {alcohol_min}")
    print(f"- Max: {alcohol_max}")
    print(f"- Mean: {alcohol_mean:.2f}")
    print(f"- Median: {alcohol_median}")
else:
    print("\n Không thể thống kê vì thiếu cột 'Alcohol'!")
print("=" * 50)

# 2.6 - Bài 1: Liệt kê các quốc gia có mức rượu cao hơn trung bình
if "Alcohol" in df.columns and "Country" in df.columns:
    high_alcohol_countries = df[df["Alcohol"] > alcohol_mean]["Country"].unique()
    print("\n Quốc gia có mức tiêu thụ rượu cao hơn trung bình:")
    print(f"- Số quốc gia: {len(high_alcohol_countries)}")
    print("- Danh sách:", ", ".join(high_alcohol_countries))
else:
    print("\n Không thể thống kê vì thiếu cột 'Alcohol' hoặc 'Country'!")
print("=" * 50)

# 2.6 - Bài 2: Quốc gia có trình độ học vấn và GDP cao hơn trung bình
if "Schooling" in df.columns and "GDP" in df.columns and "Country" in df.columns:
    education_mean = df["Schooling"].mean()
    gdp_mean = df["GDP"].mean()

    high_edu_gdp_countries = df[(df["Schooling"] > education_mean) & (df["GDP"] > gdp_mean)]["Country"].unique()
    
    print("\n Quốc gia có trình độ học vấn và GDP cao hơn trung bình:")
    print(f"- Số quốc gia: {len(high_edu_gdp_countries)}")
    print("- Danh sách:", ", ".join(high_edu_gdp_countries))
else:
    print("\n Không thể thống kê vì thiếu cột 'Schooling', 'GDP' hoặc 'Country'!")
print("=" * 50)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score


# 1. Đọc dữ liệu
df = pd.read_csv("winequality-red.csv", delimiter=";")


# Hiển thị 5 dòng đầu tiên
print("5 dòng đầu tiên:")
print(df.head(5))

# Thông tin về dữ liệu
print("\nThông tin về dữ liệu:")
print(df.info())

# Thống kê mô tả
print("\nThống kê mô tả:")
print(df.describe())


#2. Phân tích khám phá dữ liệu
#2.1 Kiểm tra giá trị thiếu
print("Số lượng giá trị thiếu trong mỗi cột:")
print(df.isnull().sum())

#2.2 Vẽ histogram cho tất cả các biến và nhận xét
df.hist(bins=20, figsize=(20, 15))
plt.tight_layout()
plt.show()
# Nhận xét: Kiểm tra phân bố của từng biến, tìm hiểu xem có bị lệch hay không


#2.3 Vẽ boxplot cho tất cả các biến và nhận xét
plt.figure(figsize=(12, 6))
sns.boxplot(data=df)
plt.xticks(rotation=90)  # Xoay tên cột nếu bị dính nhau
plt.show()
# Nhận xét: Xác định các giá trị ngoại lai.

#2.4 Vẽ Scatterplot cho "alcohol và quality"
plt.figure(figsize=(8, 5))
sns.scatterplot(x="alcohol", y="quality", data=df)
plt.title("Mối quan hệ giữa Alcohol và Quality")
plt.show()
# Nhận xét: Kiểm tra xem có mối tương quan nào giữa độ cồn và chất lượng rượu không.

#3.Tính toán ma trận tương quan và vẽ heatmap của ma trận tương quan đó
corr = df.corr()
#vẽ heatmap
plt.figure(figsize=(10, 8))  # Đặt kích thước đồ thị
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")  # Sửa lỗi colormap và format
plt.title("Heatmap của ma trận tương quan")
plt.show()


# 4.1 Xáo trộn dữ liệu (shuffle)
np.random.seed(42)  # Đặt seed để tái lập kết quả
shuffled_indices = np.random.permutation(len(df))  # Trộn chỉ mục của dữ liệu
df = df.iloc[shuffled_indices].reset_index(drop=True)  # Sắp xếp lại theo chỉ mục đã trộn

# 4.2 Chia dữ liệu với 80% huấn luyện, 20% kiểm tra
test_size = int(0.2 * len(df))  # Số lượng mẫu tập kiểm tra
train_size = len(df) - test_size  # Số lượng mẫu tập huấn luyện

train_data = df.iloc[:train_size]  # 80% dữ liệu huấn luyện
test_data = df.iloc[train_size:]  # 20% dữ liệu kiểm tra

# 4.3 Tách đặc trưng (X) và nhãn (y)
X_train = train_data.drop(columns=["quality"])  # Đặc trưng tập huấn luyện
y_train = train_data["quality"]  # Nhãn tập huấn luyện

X_test = test_data.drop(columns=["quality"])  # Đặc trưng tập kiểm tra
y_test = test_data["quality"]  # Nhãn tập kiểm tra

# Kiểm tra kích thước các tập dữ liệu
print(f"Số lượng mẫu tập huấn luyện: {X_train.shape[0]}")
print(f"Số lượng mẫu tập kiểm tra: {X_test.shape[0]}")


# 5. Chuẩn bị dữ liệu 
# 5.1 Khởi tạo bộ dữ liệu chuẩn hóa
scaler = StandardScaler()

# 5.2 Chuẩn hóa tập huấn luyện
X_train_scaled = scaler.fit_transform(X_train)

# 5.3 Chuẩn hóa tập kiểm tra
X_test_scaled = scaler.transform(X_test)

# 6. Ứng dụng mô hình hồi quy tuyến tính 
# 6.1 Khởi tạo 
model = LinearRegression()

# 6.2 Huấn luyện 
model.fit(X_train_scaled, y_train)

# 6.3 Dự đoán trên tập kiểm tra 
y_pred = model.predict(X_test_scaled)

# 7. Đánh giá mô hình 
# Tính MSE và R^2
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error (MSE): {mse:.4f}")
print(f"R-squared (R²): {r2:.4f}")
```
